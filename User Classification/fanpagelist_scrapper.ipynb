```python
from lxml import html
from lxml import etree
import requests
import re
#import json
import time
import csv

def get_tree(url):
    r = requests.get(url)
    content = r.text.encode(r.encoding).decode('utf-8', errors='ignore')
    tree = html.fromstring(content)
    print(url)
    return tree

def fetch_actor_info(item):
    subtree = etree.ElementTree(item)

    # have twitter data
    if (len(subtree.xpath("//div[@class='twitter_stats']/img/@src")) == 0): return None

    name = subtree.xpath("//span[@class='title']/text()")[0]
    # handle is a litte hacky
    raw_handle_nodes = subtree.xpath("//a[@class='clicky_ignore']/@onclick")
    handle = re.search("(?<=twitter.com/).[^']+", raw_handle_nodes[0]).group(0) if len(raw_handle_nodes) > 0 else ''
    # may or maynot exists
    category = (lambda el: el[0] if len(el) > 0 else '')(subtree.xpath("//span[@class='description'][1]/a/text()"))
    subcategory = (lambda el: el[0] if len(el) > 0 else '')(subtree.xpath("//span[@class='description'][2]/a/text()"))

    #return {'name': name, 'handle':handle, 'category': category, 'subcategory': subcategory}
    return [name, handle, category, subcategory]
    

file_name = 'fanpagelist_scrapped_data_%s.csv' % time.strftime("%d_%m_%Y")

urls = ['http://fanpagelist.com/category/brands/view/list/sort/fans/page',
    'http://fanpagelist.com/category/athletes/view/list/sort/fans/page',
    'http://fanpagelist.com/category/celebrities/view/list/sort/fans/page',
    'http://fanpagelist.com/category/sports-teams/view/list/sort/fans/page',
    'http://fanpagelist.com/category/politicians/view/list/sort/fans/page',
    'http://fanpagelist.com/category/movies/view/list/sort/fans/page',
    'http://fanpagelist.com/category/tv-shows/view/list/sort/fans/page',
    'http://fanpagelist.com/category/games/view/list/sort/fans/page',
    'http://fanpagelist.com/category/news/view/list/sort/fans/page',
    'http://fanpagelist.com/category/organizations/view/list/sort/fans/page']

for url in urls:
    
    page_number = 1
    while True:
    
        pgnted_url = url + str(page_number)
        tree = get_tree(pgnted_url)
        data = []
        items = tree.xpath("//li[@class='ranking_results']")
    
        if len(items) == 0:
            end_of_category = True
            break
        
        page_number += 1
    
        for item in items:
            actor_info = fetch_actor_info(item)
            if not actor_info: continue
            data.append(actor_info)

#         with open(file_name, 'a') as outfile:
#             json.dump(data, outfile)
            

        with open(file_name, 'a', newline='') as fp:
            file = csv.writer(fp, delimiter=',')
            file.writerows(data)
    

```
