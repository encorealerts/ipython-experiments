```python
from lxml import html
from lxml import etree
import requests
import re
import time
import csv

twitter_re = re.compile(r"http(s){0,1}:\/\/(www\.){0,1}twitter\.com")

def get_tree(url):
    r = requests.get(url)
    content = r.text.encode(r.encoding).decode('utf-8', errors='ignore')
    tree = html.fromstring(content)
    print(url)
    return tree

def fetch_actor_info(tree):

    twitter_nodes = tree.xpath("//a[@class='social_twitter']")
    
    # must have twitter data
    if len(twitter_nodes) is 0: return None
    
    twitter_href = twitter_nodes[0].attrib['href']
    
    # must have the right twitter link
    if not twitter_re.match(twitter_href): return None

    name = tree.xpath('//div[contains(concat(" ", normalize-space(@class), " "), " main ")]/div/div/h2/a/text()')[0]
    handle = re.search("(?<=twitter.com/).+", twitter_href).group(0)

    return [name, handle, 'Business', '']
    

file_name = 'socialbrandindex_scrapped_data_%s.csv' % time.strftime("%d_%m_%Y")

base_brand_url = 'http://www.socialbrandindex.com/'
base_list_url = 'http://www.socialbrandindex.com/directory?page='
brand_urls = []
page_number = 1

while True:

    pgnted_url = base_list_url + str(page_number)
    tree = get_tree(pgnted_url)
    items = tree.xpath('//a[contains(concat(" ", normalize-space(@class), " "), " btn-u-small ") and contains(concat(" ", normalize-space(@class), " "), " btn-u ")]')

    if len(items) == 0: break
        
    page_number += 1
    brand_urls += map(lambda n: base_brand_url + n.attrib['href'], items)

data = []
for url in brand_urls:
    tree = get_tree(url)
    actor_info = fetch_actor_info(tree)
    if actor_info: data.append(actor_info)
    
    if len(data) is 100:
        with open(file_name, 'a', newline='') as fp:
            file = csv.writer(fp, delimiter=',')
            file.writerows(data)
        data = []


```
